# Plotting

```python
data_CR = np.load(figpath+"square_4x4_CR.npz")
data_CZ = np.load(figpath+"square_4x4_CZ.npz")
# initialization the plot
fig, ax = plt.subplots( figsize=(7, 3.5))

ax.plot(data_CR["sigma"]*1e3, data_CR["y"], label=r'$\mathrm{CR}\, (4 \times 4)$', color = 'C0')
ax.plot(data_CZ["sigma"]*1e3, data_CZ["y"], label=r'$\mathrm{CZ}\, (4 \times 4)$', color = 'C1')
ax.plot(data_CZ["sigma"]*1e3, data_CZ["y"]**(2500/16), label=r'$\mathrm{CZ}\, (50 \times 50)$', color = 'C4')

ax.axvline(10, ls='--', color='Gray')
# ax.axvline(50, ls='--', color='Gray', label = r"JM. Kreikebaum $\it{et~al.}$")

# Legend and labels
ax.set_ylabel(f'Yield')
ax.set_xlabel('Frequency dispersion $\sigma_f$ (MHz)')
ax.set_xlim([0, 60])
ax.set_yscale('log')
# ax.set_title('Yield for collision free sample')
ax.legend(ncol=1, fontsize=12, loc='best',bbox_to_anchor=(0.6,0.5))

ax.set_ylim(1e-2, 1.2)
ax.grid()
plt.tight_layout()
figname = 'yield.pdf'
plt.savefig(figpath+figname)
```
